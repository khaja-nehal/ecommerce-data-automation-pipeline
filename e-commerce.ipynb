```python
# we are creating porfolio for our freelancing job
import pandas as pd
e_commerce_df=pd.read_csv('E Commerce Dataset.csv')
display(e_commerce_df)
e_commerce_df.info()
```

```python
# checking for inconsistance in text columns
print(e_commerce_df['PreferredLoginDevice'].unique())
print(e_commerce_df['PreferredPaymentMode'].unique())
print(e_commerce_df['PreferedOrderCat'].unique())
```

```python
# we are filling missing information
tenure_median= e_commerce_df['Tenure'].median()
e_commerce_df['Tenure']= e_commerce_df['Tenure'].fillna(tenure_median)
warehouse_median= e_commerce_df['WarehouseToHome'].median()
e_commerce_df['WarehouseToHome']= e_commerce_df['WarehouseToHome'].fillna(warehouse_median)
print("Check: New null count for Tenure:", e_commerce_df['Tenure'].isnull().sum())
print("CHeck: New null count for WarehouseToHome:", e_commerce_df['WarehouseToHome'].isnull().sum())
```

```python
# we are filling missing value of hourspendonapp with 0
e_commerce_df['HourSpendOnApp']= e_commerce_df['HourSpendOnApp'].fillna(0)
# we are doing the same thing for coupneusers
e_commerce_df['CouponUsed']= e_commerce_df['CouponUsed'].fillna(0)
# we are filling the ordercount with minimum order
e_commerce_df['OrderCount']= e_commerce_df['OrderCount'].fillna(1)
# we are filling DaySinceLastOrder with median
DaySinceLastOrder_median= e_commerce_df['DaySinceLastOrder'].median()
e_commerce_df['DaySinceLastOrder']= e_commerce_df['DaySinceLastOrder'].fillna(DaySinceLastOrder_median)
print("Check: new null count:", e_commerce_df['HourSpendOnApp'].isnull().sum())
print("check: new null count:", e_commerce_df['CouponUsed'].isnull().sum())
print("check: new null count:", e_commerce_df['OrderCount'].isnull().sum())
print("check: new null count:", e_commerce_df['DaySinceLastOrder'].isnull().sum())
```

```python
# we are replacing duplicate items to one item
e_commerce_df['PreferredLoginDevice']= e_commerce_df['PreferredLoginDevice'].replace('Phone', 'Mobile Phone')
e_commerce_df['PreferredPaymentMode']= e_commerce_df['PreferredPaymentMode'].replace('CC','Credit Card')
e_commerce_df['PreferredPaymentMode']= e_commerce_df['PreferredPaymentMode'].replace('COD','Cash on Delivery')
e_commerce_df['PreferedOrderCat']= e_commerce_df['PreferedOrderCat'].replace('Mobile Phone', 'Mobile')
print(e_commerce_df['PreferredLoginDevice'].unique())
print(e_commerce_df['PreferredPaymentMode'].unique())
print(e_commerce_df['PreferedOrderCat'].unique())
```

```python
# This shows the statistics (Min, Max, Average) for your numbers
print(e_commerce_df[['Tenure', 'WarehouseToHome', 'OrderCount', 'CashbackAmount']].describe())
```
